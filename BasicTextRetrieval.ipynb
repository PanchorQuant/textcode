## <b> Step 1: Text Preprocessing </b>
Trong các bài toán Máy học liên quan đến dữ liệu văn bản (text), tiền xử lý văn bản (text preprocessing) là một kỹ thuật cực kì quan trọng, góp phần làm cải thiện khả năng biểu diễn, phân tích ngôn ngữ và trên hết là gia tăng hiệu suất mô hình. Một trong số các kỹ thuật tiền xử lý văn bản phổ biến gồm có:


(a) <b> Chuyển chữ viết thường (lowercasing): </b> Vì máy tính coi hai chữ cái 'a' và 'A' là hai ký tự khác nhau mặc dù có cùng ý nghĩa, điều này làm gia tăng độ phức tạp cũng như tài nguyên lưu trữ khi biểu diễn văn bản. Lowercasing là kỹ thuật chuyển toàn bộ các chữ cái trong đoạn văn bản thành kiểu chữ viết thường, sử dụng lowercasing sẽ phần nào giúp ta tránh được trường hợp kể trên. <br>
Ví dụ: <b> "Hello we're AIVN"   --->   "hello we're aivn"

(b) <b> Chuyển chữ viết hoa (uppercasing) </b> : Với mục đích tương tự như kỹ thuật lowercasing, ở kỹ thuật uppercasing các chữ cái có trong văn bản sẽ được chuyển toàn bộ sang kiểu chữ viết hoa. <br>
Ví dụ: <b> "Hello we're AIVN" --> "HELLO WE'RE AIVN" </b>

Mục đích chính của việc loại bỏ các url đó là vì chúng không thật sự biểu diễn đúng ngôn ngữ tự nhiên, vì vậy nhiều khả năng sẽ không chứa những thông tin quan trọng trong việc biểu diễn văn bản.

```python
!pip install autocorrect
!pip install stopwords
!pip install tqdm

import math
import os
import string
import nltk
import re
import spacy
import numpy as np
import pandas as pd
import unicodedata

from autocorrect import Speller
from nltk.stem import PorterStemmer
from tqdm import tqdm
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
class TextPreprocessor:
    def __init__(self):
        self.stopwords = stopwords
        self.stemmer = PorterStemmer()
        self.spelling_correction = Speller(lang= 'en')
    
    def lowercase(self, sample):
        return sample.lower()
    
    def uppercase(self, sample):
        return sample.upper()

    def remove_email(self, sample):
        '''Remove email'''
        return re.sub(r'\S*@\S*\s?', '', sample)

    def remove_url(self, sample):
        '''Remove url link'''
        return re.sub(r"http\S+", "", sample)

    def remove_tags(self, sample):
        '''Remove html tags'''
        soup = BeautifulSoup(sample, 'html.parser')
        return (soup.get_text(strip= True))

    def remove_punct(self, sample):
        '''Remove punctuation'''
        sample = sample.strip()
        chars = sample.split(" ")
        result = []
        for char in chars:
            char = char.translate(str.maketrans('', '', string.punctuation))
            result.append(char)
        return ' '.join(result)

    def remove_stopwords(self, sample):
        '''Remove stopwords''' 
        sample = sample.strip()
        chars = sample.split(" ")
        result = []
        for char in chars:
            if char not in self.stopwords.words('english'):
                result.append(char)
        return ' '.join(result)

    def remove_freqs(self, sample):
        '''Remove frequence'''
        sample = sample.strip()
        chars = sample.split(' ')
        freq = {}
        result = []
        for char in chars:
            if(char in freq):
                freq[char] += 1
            else:
                freq[char] = 1
        for char in chars:
            if(freq[char] == 1):
                result.append(char)
        return ' '.join(result)

    def word_correct(self, sample):
        '''Spelling correction'''
        sample = sample.strip()
        chars = sample.split(' ')
        result = []
        for char in chars:
            result.append(self.spelling_correction(char))
        return ' '.join(result)

    def stemming(self, sample):
        '''Stemming'''
        sample = sample.strip()
        chars = sample.split(' ')
        result = []
        for char in chars:
            result.append(self.stemmer.stem(char))
        return ' '.join(result)

    def lemmatize(self, sample):
        '''Lemmatization'''
        nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
        sample = sample.strip()
        chars = nlp((sample))
        result = []
        for char in chars:
            result.append(char.lemma_)
        return ' '.join(result)
    
    def preprocess_text(self, text, lowercase= False, uppercase= False, remove_email = False, remove_punct= False, remove_stopwords= False, remove_freqs=  False , stemming= True, lemmatize= False, remove_url= True, remove_tags= True, spelling_correct= True):
        if (lowercase & uppercase) :
            raise Exception("Cannot do both lowercasing and uppercasing. Please specify only one of them.")
            return None
        if (stemming & lemmatize):
            raise Exception("Cannot do both stemming and lemmatization. Please specify only one of them. ")
            return None

        if(lowercase):
            text = self.lowercase(text)
            # print(text)

        if(uppercase):
            text = self.uppercase(text)
            # print(text)

        if(remove_email):
            text = self.remove_email(text)

        if(remove_url):
            text = self.remove_url(text)
            # print(text)

        if(remove_tags):
            text = self.remove_tags(text)
            # print(text)
        
        if(remove_punct):
            text = self.remove_punct(text)
            # print(text)

        if(remove_stopwords):
            text = self.remove_stopwords(text)
            # print(text)

        if(remove_freqs):
            text = self.remove_freqs(text)
            # print(text)

        if(stemming):
            text = self.stemming(text)
            # print(text)

        if(lemmatize):
            text = self.lemmatize(text)
            # print(text)

        if(spelling_correct):
            text = self.word_correct(text)
            # print(text)
            
        return text.replace('\n', ' ')
```

## **Normalized text**

```python
def normalize_text(text: str) -> str:
    processor = TextPreprocessor()
    normalized_text = processor.preprocess_text(text, lowercase = True)
    normalized_text = unicodedata.normalize('NFKC', normalized_text)
    normalized_text = processor.preprocess_text(text, remove_email = True, remove_url = True, remove_punct = True, remove_stopwords = True)

    return normalized_text
```

## **Xây dựng bộ từ vựng (vocab):** 
Từ bộ ngữ liệu cho trước, ta duyệt qua nội dung của từng tài liệu, thực hiện chuẩn hóa văn bản và sử dụng kỹ thuật Tokenization để tách thành các token. Với mỗi token nhận được, kiểm tra nếu token hiện tại không tồn tại trong bộ từ vựng thì ta sẽ thêm nó vào.
image.png

```python
doc_lists = []
vocab = []
filenames = os.listdir(path1)

for i in tqdm(range(len(filenames) // 200)):
    filename = filenames[i]
    filepath = os.path.join(path1, filename)
    with open(filepath, 'r', encoding='utf8') as f:
        lines = list(filter(None, f.read().splitlines()))
        title = unicodedata.normalize('NFKC', lines[0].strip())
        article = ' '.join(lines[1:]).strip()
        article = normalize_text(article)
        if article == '':
            continue
        else:
            if (title, article) not in doc_lists:
              doc_lists.append((title, article)) 
            tokens = list(filter(None, article.split(' ')))
            for token in tokens:
              if token not in vocab:
                vocab.append(token)
```

## **Xây dựng hàm vector hóa văn bản:** 
Vì máy tính không thể sử dụng chuỗi văn bản để thực hiện tính toán, ta cần biểu diễn lại các văn bản dưới dạng vector. Trong project này, ta xây dựng một hàm vectorize() với tham số đầu vào là một chuỗi văn bản s và bộ từ vựng vocab, sau đó thực hiện vector hóa văn bản sử dụng kỹ thuật bag-of-words (các bạn có thể tham khảo lý thuyết về bag-of-words tại [đây](https://www.mygreatlearning.com/blog/bag-of-words/)) dựa trên bộ từ vựng cho trước, kết quả trả về sẽ là vector bag-of-words của văn bản đầu vào.

```python
def vectorize(text: str, vocab: list) -> np.ndarray:
    normalized_text = normalize_text(text)
    vec = []
    for word in vocab:
        vec.append(normalized_text.count(word))

    return np.array(vec)
```

## **Xây dựng ma trận document-term:**
Với các tài liệu trong bộ ngữ liệu, ta cần lưu trữ dạng biểu diễn vector của chúng trong một cấu trúc được gọi là ma trận document-term. Trong ma trận này, mỗi hàng sẽ đại diện cho một tài liệu và mỗi cột sẽ đại diện cho mỗi từ (term) trong bộ từ vựng. Như vậy, giả sử với một bộ ngữ liệu gồm có ba tài liệu văn bản có nội dung như sau: 
* doc1 = "Học sách học AI!"
* doc2 = "Sách Học Máy."
* doc3 = "Người ấy là ai?"

Sau khi thực hiện chuẩn hóa văn bản và tokenization, ta thu thập các token độc nhất trong toàn bộ bộ ngữ liệu để lập thành bộ từ vựng gồm [’học’, ’sách’, ’ai’, ’máy’, ’người’, ’ấy’, ’là’]. Dựa vào danh sách từ vựng có được, tính tần suất xuất hiện của mỗi từ trong bộ từ vựng với từng văn bản đã được tokenize để từ đó có được dạng vector biểu diễn của chúng. Trong ma trận document-term, ba vector này chính là 3 hàng của ma trận tượng trưng cho 3 tài liệu và các giá trị trong vector sẽ tượng trưng cho các cột đại diện cho các từ trong bộ từ vựng.

```python
doc_term_matrix = {}
for (title, article) in tqdm(doc_lists):
  vec = vectorize(article, vocab)
  doc_term_matrix[(title, article)] = vec
```

## **Xây dựng hàm tính độ tương đồng giữa hai vector:** 

Để có thể tìm được các tài liệu có liên quan đến câu truy vấn, ta có thể sử dụng các công thức được dùng để đo sự tương đồng giữa hai vector (ở đây sẽ hiểu là vector tài liệu và vector câu truy vấn), từ đó xây dựng một hàm **distance()** nhận đầu vào là hai vector có cùng kích thước, sau đó trả về một giá trị là điểm đại diện cho độ tương đồng của vector này

**Cosine Similarity**

$$cosine\_similarity(a, b) = \frac{a ⋅ b}{\|a\|\|b\|} = \frac{\displaystyle\sum_{i = 1}^{N}a_ib_i}{\sqrt{\displaystyle\sum_{i = 1}^{N}a_i^2}\sqrt{\displaystyle\sum_{i = 1}^{N}b_i^2}}$$

```python
def distance(a: np.ndarray, b: np.ndarray) -> np.float64:
  numerator = np.dot(a, b)
  denominator = np.linalg.norm(a) * np.linalg.norm(b)
  
  return numerator / denominator
```

## **Thực hiện truy vấn:** 
Cuối cùng, dựa trên các thành phần đã xây dựng ở các bước trên, ta sẽ bắt đầu thực hiện tìm kiếm các tài liệu có liên quan đến một chuỗi văn bản câu truy vấn cho trước, các bước thực hiện như sau:

```python
def ranking(query: str, doc_term_matrix: dict, print_top_10: bool = True) -> list:
    query_vec = vectorize(query, vocab)
    rankings = []
    i = 1
    for doc_info, vec in doc_term_matrix.items():
        score = distance(query_vec, vec)
        rankings.append((score, (doc_info[0])))
        i += 1
    rankings.sort(reverse=True)

    if print_top_10 == True:
        for rank in rankings[:10]:
            print(rank)

    return rankings
```
